## 4.1: Build plots from the ground-up using plot scaffolding

```python
# set matpltolob to draw plots inline
%matplotlib inline

# import matplotlib pyplot
import matplotlib.pyplot as plt

# import numpy
import numpy as np 

# import pandas
import pandas as pd
```

```python
# Data
#  Bank and Marketing Data from UCI
# *from https://archive.ics.uci.edu/ml/datasets/bank+marketing

# import dataset
data = pd.read_csv("bank.csv", sep=';')
```

```python
data.head()
```

```python
plt.scatter(data['age'], data['balance'])
plt.show()
```

```python
plt.scatter(data['age'], data['balance'])
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
plt.scatter(data['age'], data['balance'])
plt.xlabel('Age')
plt.ylabel('Balance')
plt.title("Balance by Age")
plt.show()
```

## 3.2: Building custom plots using figures

```python
fig = plt.figure()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.scatter(data['age'], data['balance'], color='darkgreen')
plt.show()
```

```python
fig = plt.figure(figsize = [12, 6])
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.scatter(data['age'], data['balance'], color='darkgreen')
plt.show()
```

```python
fig = plt.figure(figsize = [12, 6], facecolor = 'y')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.scatter(data['age'], data['balance'], color='darkgreen')
plt.show()
```

```python
fig = plt.figure(figsize = [12, 6], facecolor = 'y', edgecolor='black')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.scatter(data['age'], data['balance'], color='darkgreen')
plt.show()
```

## 3.3: Customizing plot axes

```python
fig = plt.figure(figsize = [12, 6], facecolor = 'y', edgecolor='black')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
plt.show()
```

```python
fig = plt.figure(figsize = [12, 6], facecolor = 'y', edgecolor='black')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.set_xlabel('Customer')
ax.set_ylabel('Balance')
plt.show()
```

```python
fig = plt.figure(figsize = [12, 6], facecolor = 'y', edgecolor='black')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.set_xlabel('Customer')
ax.set_ylabel('Balance')
plt.show()
```

```python
fig = plt.figure(figsize = [12, 6], facecolor = 'y', edgecolor='black')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.set_xlabel('Customer',  fontsize=18)
ax.set_ylabel('Balance',  fontsize=18)
plt.show()
```

```python
fig = plt.figure(figsize = [12, 6], facecolor = 'y', edgecolor='black')
ax = fig.add_subplot(111)
ax.plot(data.balance, color='lightblue', linewidth=3)
ax.set_xlabel('Customer',  fontsize=18)
ax.set_ylabel('Balance',  fontsize=18)
ax.xaxis.label.set_color('red')
ax.yaxis.label.set_color('green')
plt.show()
```
