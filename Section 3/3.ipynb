## 3.2: Adding grids and customizing ticks

```python
data.balance.plot.hist()
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.show()
```

```python
data.balance.plot.hist(grid=True)
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.show()
```

```python
data.balance.plot.hist(grid=True, bins=15, rwidth=0.9,
                   color='#607c8e')
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.show()
```

```python
data.balance.plot.hist(grid=True, bins=15, rwidth=0.9,
                   color='#607c8e')
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.grid(color='b', linestyle='-', linewidth=0.5)
plt.show()
```

```python
data.balance.plot.hist(grid=True, bins=15, rwidth=0.9,
                   color='#607c8e')
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.grid(color='b', linestyle='-', linewidth=1)
plt.xticks()
plt.show()
```

```python
data.balance.plot.hist(grid=True, bins=15, rwidth=0.9,
                   color='#607c8e')
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.grid(color='b', linestyle='-', linewidth=1)
plt.xticks([])
plt.show()
```

```python
data.balance.plot.hist(grid=True, bins=15, rwidth=0.9,
                   color='#607c8e')
plt.ylabel('Customer')
plt.xlabel('Balance')
plt.title('Customer Balance Distribution')
plt.grid(color='b', linestyle='-', linewidth=1)
plt.xticks(np.arange(0, 1000000, step=100000))
plt.show()
```

## 3.5: Plot Annotation

```python
plt.figure(1, figsize=(3,3))
ax = plt.subplot(111)

ax.annotate("Here",
            xy=(0.2, 0.2), xycoords='data',
            xytext=(0.8, 0.8), textcoords='data',
            size=20, va="center", ha="center",
            arrowprops=dict(arrowstyle="simple",
                            connectionstyle="arc3,rad=-0.2"), 
            )

plt.show()
```

```python
fig, ax = plt.subplots()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
ax.set_ylim(-2, 2)
plt.show()
```
