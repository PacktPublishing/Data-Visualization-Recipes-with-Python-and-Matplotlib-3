```python
# set matpltolob to draw plots inline
%matplotlib inline

# import matplotlib pyplot
import matplotlib.pyplot as plt

# import numpy
import numpy as np 

# import pandas
import pandas as pd
```

####################################################

```python
# Online Retail Data Set
# *from https://archive.ics.uci.edu/ml/datasets/online+retail

# import dataset (excel file)
data = pd.read_excel("online-retail.xlsx")
```

```python
# Explore Data
data.head()
```

```python
data.describe()
```

```python
# Plotting a scatter plot

plt.figure(figsize=(10,6))
plt.scatter(data['UnitPrice'], data['Quantity'])
plt.xlabel('Unit Price')
plt.ylabel('Quantity')
plt.show()
```

```python
# Plotting another scatter plot with different columns data

plt.figure(figsize=(10,6))
plt.scatter(data['CustomerID'], data['Quantity'])
plt.xlabel('Customer ID')
plt.ylabel('Quantity')
plt.show()
```

```python
# Plotting another scatter plot with different columns data

plt.figure(figsize=(10,6))
plt.scatter(data['CustomerID'], data['Quantity'], s=data['UnitPrice'])
plt.xlabel('Customer ID')
plt.ylabel('Quantity')
plt.show()
```
